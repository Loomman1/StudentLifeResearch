```python
#Цель исследования - изучение зависимости итоговых оценок студентов от их стиля жизни, условий,
#и других факторов, влияющих на процесс обучения.


#Student Performance & Behavior Dataset
#This dataset is real data of 5,000 records collected from a private learning provider.
# The dataset includes key attributes necessary for exploring patterns, correlations, and insights
#related to academic performance.

# Columns:

# # Student_ID: Unique identifier for each student.
# # First_Name: Student’s first name.
# # Last_Name: Student’s last name.
# # Email: Contact email (can be anonymized).
# # Gender: Male, Female, Other.
# # Age: The age of the student.
# # Department: Student's department (e.g., CS, Engineering, Business).
# # Attendance (%): Attendance percentage (0-100%).
# # Midterm_Score: Midterm exam score (out of 100).
# # Final_Score: Final exam score (out of 100).
# # Assignments_Avg: Average score of all assignments (out of 100).
# # Quizzes_Avg: Average quiz scores (out of 100).
# # Participation_Score: Score based on class participation (0-10).
# # Projects_Score: Project evaluation score (out of 100).
# # Total_Score: Weighted sum of all grades.
# # Grade: Letter grade (A, B, C, D, F).
# # Study_Hours_per_Week: Average study hours per week.
# # Extracurricular_Activities: Whether the student participates in extracurriculars (Yes/No).
# # Internet_Access_at_Home: Does the student have access to the internet at home? (Yes/No).
# # Parent_Education_Level: Highest education level of parents (None, High School, Bachelor's, Master's, PhD).
# # Family_Income_Level: Low, Medium, High.
# # Stress_Level (1-10): Self-reported stress level (1: Low, 10: High).
# # Sleep_Hours_per_Night: Average hours of sleep per night.



# Импортирование основных библиотек
import pandas as pd              # Для работы с данными в формате DataFrame
import numpy as np               # Для работы с числовыми операциями
import matplotlib.pyplot as plt   # Для визуализации данных
import seaborn as sns            # Для более красивой визуализации по сравнению с matplotlib
! pip install phik
import phik
from phik.report import plot_correlation_matrix
from phik import report

sns.set(style='whitegrid')
! gdown --id 1CB0tMF7CqLdLVp7p55R7fjuDpc5OirdW
#link='https://drive.google.com/drive/folders/1KyVLjhKrq-RSVeSvRTqm44t5tbmQ2rLv?usp=drive_link' #ссылка на папкугугл диск
# dataset = pd.read_csv('E:/Python projects/data/students_dataset/Students_Grading_Dataset.csv')
df = pd.read_csv('/content/Students_Grading_Dataset.csv', encoding='cp1251')
print(df.head())
```

```python
female_scores=df[df['Gender']=='Female']
plt1=female_scores.groupby(['Grade','Gender'])['Final_Score'].mean().plot(legend=True)
male_scores=df[df['Gender']=='Male']
plt1=male_scores.groupby(['Grade','Gender'])['Final_Score'].mean().plot(legend=True)
plt1.set_title('Male/Female grades')     # Заголовок для второго графика
plt1.set_xlabel('Grades')              # Подпись оси X
plt1.set_ylabel('Final_score')         # Подпись оси Y
```

```python

# Группировка данных по 'Grade' для женских оценок
female_scores = df[df['Gender'] == 'Female']
female_means = female_scores.groupby('Grade')['Final_Score'].mean()

# Группировка данных по 'Grade' для мужских оценок
male_scores = df[df['Gender'] == 'Male']
male_means = male_scores.groupby('Grade')['Final_Score'].mean()

# Построение графиков
plt.figure(figsize=(10, 6))

female_means.plot(kind='bar', position=0, width=0.4, label='Female', color='pink')
male_means.plot(kind='bar', position=1, width=0.4, label='Male', color='lightblue')

plt.title('Average Final Score by Grade and Gender')
plt.xlabel('Grade')
plt.ylabel('Average Final Score')
plt.xticks(rotation=0)  # Убираем наклон для меток по оси X
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Группировка данных по 'Grade' для женских оценок
female_scores = df[df['Gender'] == 'Female']
female_means = female_scores.groupby('Grade')['Attendance (%)'].mean()

# Группировка данных по 'Grade' для мужских оценок
male_scores = df[df['Gender'] == 'Male']
male_means = male_scores.groupby('Grade')['Attendance (%)'].mean()

# Построение графиков
plt.figure(figsize=(10, 6))

female_means.plot(kind='bar', position=0, width=0.4, label='Female', color='pink')
male_means.plot(kind='bar', position=1, width=0.4, label='Male', color='lightblue')

plt.title('Average Grade by Attendance (%) and Gender')
plt.xlabel('Grade')
plt.ylabel('Average Attendance (%)')
plt.xticks(rotation=0)  # Убираем наклон для меток по оси X
plt.legend()
plt.tight_layout()
plt.show()
```
